```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import pickle
import numpy as np
import pandas as pd
```

```python
def init_process(fin,fout):
    outfile = open(fout,'a')
    with open(fin, buffering=5572, encoding='latin-1') as f:
        try:
            for line in f:
                line = line.replace('"','')
                initial_polarity = line.split(',')[0]
                if initial_polarity == 'ham':
                    initial_polarity = [1,0]
                elif initial_polarity == 'spam':
                    initial_polarity = [0,1]
                SMS = line.split(',')[1]
                outline = str(initial_polarity)+':::'+SMS
                outfile.write(outline + "\n")
        except Exception as e:
            print(str(e))
    outfile.close()
```

```python
init_process('spam.csv','Data_set.csv')
```

```python
lemmatizer = WordNetLemmatizer()
def create_lexicon(fin):
    lexicon = []
    with open(fin, 'r', buffering=5572, encoding='latin-1') as f:
        try:
            counter = 1
            content = ''
            for line in f:
                counter += 1
                tweet = line.split(':::')[1] # get the split right entry which is the statement 
                #print(tweet)
                content += ' '+tweet
                words = word_tokenize(content.lower()) # get all words
                words = [lemmatizer.lemmatize(i) for i in words] # lemmatize all words
                lexicon = list(set(lexicon + words)) # add all words in one lexicon
                #print(counter, len(lexicon))

        except Exception as e:
            print(str(e))

    with open('lexicon.pickle','wb') as f:
        pickle.dump(lexicon,f) # create a pickle 
```

```python
create_lexicon('Data_set.csv')
```

```python
def shuffle_data(fin):
    df = pd.read_csv(fin, error_bad_lines=False)
    df = df.iloc[np.random.permutation(len(df))]
    print(df.head())
    df.to_csv('train_set_shuffled.csv', index=False)
    
shuffle_data('Data_set.csv')
```

```python
def sample_handling(fin,fout,lexicon_pickle):
    
    with open(lexicon_pickle,'rb') as f:
        lexicon = pickle.load(f)
    outfile = open(fout,'a')

    with open(fin, buffering=5574, encoding='latin-1') as f:
        try:
            counter = 0
            for line in f:
                counter +=1
                
                label = line.split(':::')[0]
                SMS = line.split(':::')[1]
                
                current_words = word_tokenize(SMS.lower())
                current_words = [lemmatizer.lemmatize(i) for i in current_words]
                
                features = np.zeros(len(lexicon))
                
                for word in current_words:
                    if word.lower() in lexicon:
                        index_value = lexicon.index(word.lower())
                        features[index_value] += 1

                features = list(features)
                outline = str(features)+'::'+str(label)+'\n'
                outfile.write(outline)
                
        except Exception as e:
            print(str(e))
            
    print(counter)
```

```python
sample_handling('train_set_shuffled.csv','processed.csv','lexicon.pickle')
```
